# 2009 / 2010

Progress points for the premier league season 2016/2017.

```python
%config InlineBackend.figure_formats = ['svg']
from proggyleg import proggyleg
year = 2009
```

## Cumulative Points

The sum of points up to a given game played.

```python
proggyleg.autoplot(year, league="E0", which="cumulative");
```

## Extrapolated Points

The sum of points assuming each team continues to score points at the same rate as they have so far.

```python
proggyleg.autoplot(year, league="E0", which="extrapolated");
```

## Position

```python
proggyleg.autoplot(year, league="E0", which="position");
```

## Rolling Form

Exponential weighted moving average of points.

```python
proggyleg.autoplot(year, league="E0", which="form");
```
